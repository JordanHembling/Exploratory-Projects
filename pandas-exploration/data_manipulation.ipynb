#### Provided mask helpers 


```python
df.loc[df.age.between(0,35)]
```

### Views vs Copy

```python
df2 = df.copy()
```

```python
df2["age"][0] = 99
```

```python
df2.head()
```

```python
df2.loc[df2.index == 0, "age"] = 110

df2.head(1)
```

```python
df2[df2.age==110]["chol"] = 500
df2
```

Note: loc will mute data but using a mask directly on the data frame will not.
