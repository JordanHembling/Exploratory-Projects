```python
base_site = "https://itunes.apple.com/search"
```

```python
import requests
import json
```

```python
url = base_site + "?terms=the+beatles&country=us"
requests.get(url)
```

```python
r = requests.get(base_site, params = {"term": "the beatles", "country": "us"})
r.status_code
```

```python
info = r.json()
print(json.dumps(info, indent=4))
```

```python
info.keys()
```

```python
print(json.dumps(info['results'][1], indent=4))
```

```python
info["resultCount"]
```

```python
print(json.dumps(info['results'].['artistName']))
```

### Structuring and exporting data with pandas

```python
import pandas as pd
```

```python
songs_df = pd.DataFrame(info["results"])
songs_df
```

```python
songs_df.to_csv("songs_info.csv")
```

```python
songs_df.to_excel("songs_info.xlsx")
```
