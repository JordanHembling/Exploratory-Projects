### Incorporating Parameters in the GET requests

```python
param_url = access_url + "&symbols=USD,GBP"
param_url
```

```python
response = requests.get(param_url)
response
```

```python
data = response.json() 
data
```

```python
param_url = access_url + "&symbols=GBP" + "&" + "base=USD"
param_url
```

```python
data = requests.get(param_url).json()
data
```
